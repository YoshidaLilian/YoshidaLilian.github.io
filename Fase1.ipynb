Crie um algoritmo em Python que peça as 4 notas bimestrais e mostre a média.

```
print("Esse é um programa que cacula a média das suas provas bimestrais.")
p1 = float(input("Digite a nota da sua prova 1: "))
p2 = float(input("Digite a nota da sua prova 2: "))
p3 = float(input("Digite a nota da sua prova 3: "))
p4 = float(input("Digite a nota da sua prova 4: "))
print(f"A média das provas é {((p1+p2+p3+p4)/4):.2f}.")
```

Crie um algoritmo em Python que peça o raio de um círculo, calcule e mostre sua área.

```
print("Esse programa calcula a área de um círculo, a partir do raio informado.")
raio = float(input("Digite o raio em cm: "))
area = 3.14*(raio**2)
print(f"A área do círculo é {area:.2f} cm².")
```

Crie um algoritmo em Python que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.

```
print("Esse programa calcula a área de um quadrado e retorna o dobro desse valor para o usuário.")
lado = float(input("Digite o tamanho do lado do quadrado, em cm: "))
area = lado**2
print(f"O dobro da área do quadrado é {2*area:.2f} cm².")
```

Crie um algoritmo em Python que peça a temperatura em graus Celsius, transforme e mostre em graus Farenheit.

```
print("Esse programa converte a temperatura de graus Celsius para Fahrenheit.")
c = float(input("Digite a temperatura em graus Celsius: "))
f = (c*(9/5))+32
print(f"A temperatura em graus Fahrenheit é {f:.2f} °F.")
```

* a soma do dobro do primeiro com metade do segundo .
* a soma do triplo do primeiro com o terceiro.
* o terceiro elevado ao cubo.
