Faça um programa para a leitura de duas notas parciais de um aluno. O programa deve calcular a média alcançada por aluno e apresentar:
* "Aprovado", se a média alcançada for maior ou igual a sete;
* "Reprovado", se a média for menor do que sete;
* "Aprovado com Distinção", se a média for igual a 10.

```
nota1 = float(input("Digite a primeira nota parcial: "))
nota2 = float(input("Digite a segunda nota parcial: "))
media = (nota1+nota2)/2

if media == 10:
  print("Aprovado com distinção :D")
elif media >= 7:
  print("Aprovado :)")
else:
  print("Reprovado :(")
```

* salários entre 680,00 e 800,00 : aumento de 15%<br>

* o salário antes do reajuste;
* o percentual de aumento aplicado;
* o valor do aumento;
* o novo salário, após o aumento.
